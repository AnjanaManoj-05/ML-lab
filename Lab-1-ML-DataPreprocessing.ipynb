```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv("/Dataset of Diabetes .csv")
df2=pd.read_csv("/adult.csv")
```

```python
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
df2.replace('?', np.nan, inplace=True)

# Fill categorical with mode
for col in df2.select_dtypes(include='object').columns:
    df2[col].fillna(df2[col].mode()[0], inplace=True)

# Fill numeric with mean
df2.fillna(df2.mean(numeric_only=True), inplace=True)
```

```python
df1
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer



categorical_cols = ['Gender', 'CLASS']


ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)],
    remainder='passthrough'
)

# 4. Fit and Transform
df1_encoded_array = ct.fit_transform(df1)

# 5. Correctly create DataFrame
df1_encoded = pd.DataFrame(df1_encoded_array, columns=ct.get_feature_names_out())
print(df1_encoded.head())
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer



categorical_cols = ['workclass', 'education','marital-status'	,'occupation','relationship','race','gender','native-country','income']


ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)],
    remainder='passthrough'
)

# 4. Fit and Transform
df2_encoded_array = ct.fit_transform(df2)

# 5. Correctly create DataFrame
df2_encoded = pd.DataFrame(df2_encoded_array, columns=ct.get_feature_names_out())
print(df2_encoded.head())
```
